```python
df['Date_of_Journey']= pd.to_datetime(df['Date_of_Journey'])
```

```python
df.info()
```

```python
df['Total_Stops'].unique()
```

```python
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
df['Arrival_Hour'] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival_Minute'] = df['Arrival_Time'].str.split(':').str[1]
```

```python
df['Arrival_Hour']=df['Arrival_Hour'].astype(int)
df['Arrival_Minute']=df['Arrival_Minute'].astype(int)
```

```python
df.info()
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df['Dept_Hour'] = df['Dep_Time'].str.split(':').str[0]
df['Dept_Minute'] = df['Dep_Time'].str.split(':').str[1]

```

```python
df['Dept_Hour']=df['Dept_Hour'].astype(int)
df['Dept_Minute']=df['Dept_Minute'].astype(int)
```

```python
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
df['Total_Stops'] = df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,np.nan:1})
```

```python
df[df['Total_Stops'].isnull()]
```

```python
df['Duration_Hour']= df['Duration'].str.split(' ').str[0].str.split('h').str[0]
```

```python
df['Duration_Min']= df['Duration'].str.split(' ').str[1]
```

```python
df['Duration_Min'].isnull().sum()
```

```python
df['Duration_Min'] = df['Duration_Min'].fillna('1m')
```

```python
df['Duration_Min']= df['Duration_Min'].str.split('m').str[0]
```

```python
df['Journey_Duration'] = df['Duration_Hour'] * 60 + df['Duration_Min']
```

```python
df['Duration_Min']= df['Duration_Min'].astype(int)
df['Duration_Hour']= df['Duration_Hour'].astype(int)
```

```python
df['Duration_Hour'].unique()
```

```python
df = df[~df['Duration_Hour'].astype(str).str.contains('m', na=False)]
```

```python
df.info()
```

```python
df['Journey_Duration']= df['Journey_Duration'].astype(object)
```

```python
df.drop(['Duration_Hour','Duration_Min','Duration'],axis=1,inplace=True)
```

```python
df['Journey_Duration']= df['Journey_Duration'].astype(str) + 'm'
```
