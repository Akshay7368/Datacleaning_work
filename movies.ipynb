```python
##Changing Datatypes of Columns

df['budget'] = df['budget'].astype('Int64')
df['gross'] = df['gross'].astype('Int64')
```

```python
#Comparing the year columns 
df[['year','released','released_year']]
```

```python
#r'(\d{4}) It is a Regex Function which will exactly extract the year from the released column
df['released_year'] = df['released'].str.extract((r'(\d{4})'))
```

```python
df['released_year']= df['released_year'].astype('Int64')
```

```python
#Comparing again
mismatch = df[df['year'] != df['released_year']]
print(mismatch[['name', 'year', 'released', 'released_year']].head(10))
```

```python
#year column shows the correct year , So droping the other two columns
df.drop(columns=['released','released_year'],axis=1,inplace=True)
```

```python
#renaming
df.rename(columns={'year':'released_year'})
```

```python
#Sorting the columns according to our need
df = df.sort_values(by='gross',inplace=False,ascending=False)
```

```python
plt.scatter(x=df['budget'],y= df['gross'])
plt.title("Gross Vs Budget")
plt.xlabel("Gross")
plt.ylabel("Budget")
plt.grid(True)
plt.show()
```

```python
sns.regplot(x='budget',y='gross',data=df,line_kws={"color": "black"})
```

```python
#Correlation in python has (Pearson, kendall, spearman)
#Correlation measures the strength and direction of the relationship between two variables
# High Correlation for budget and gross

co_relation= df.corr(method='pearson',numeric_only=True)
```

```python
sns.heatmap(co_relation,annot=True)
plt.title("Correlation of the Dataset")
```

```python
df_numerized = df
for col in df_numerized.columns:
    if(df_numerized[col].dtype=='object'):
       df_numerized[col] =  df_numerized[col].astype('category')
       df_numerized[col] =  df_numerized[col].cat.codes
df_numerized
```

```python
co_relation= df_numerized.corr(method='pearson',numeric_only=True)
plt.figure(figsize=(12,10))
sns.heatmap(co_relation,annot=True)
plt.title("Correlation of the Dataset")
```

```python
corr_mat = df_numerized.corr()
corr_pair = corr_mat.unstack()
corr_pair
```

```python
sorted_val = corr_pair.sort_values()
high_corr = sorted_val[(sorted_val) > 0.5]
high_corr
```
